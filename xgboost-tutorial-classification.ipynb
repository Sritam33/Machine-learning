```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import xgboost as xgb
```

# Load and Prepare data


```python
df = pd.read_csv("/kaggle/input/docspot/datasets_228_482_diabetes.csv")
```

```python
df
```

```python
# split data into x and y
x = df.iloc[:,:-1].values #select all rows and columns except last one
y = df.iloc[:,-1].values   # select all rows and only last column
```

```python
from sklearn.model_selection import train_test_split # split the dataset for training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state=5)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from xgboost import XGBClassifier
# fit model no training data
model = XGBClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred) * 100
print("accuracy of the logistic regression model:",accuracy)
```
